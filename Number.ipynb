```
import tensorflow as tf
```

```
tf.__version__
```

```
minst=tf.keras.datasets.mnist #28*28 images of handwriting digits (0-9)
```

```
(x_train,y_train),(x_test,y_test)= minst.load_data()
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(x_train[0])
```

```
plt.imshow(x_train[1])
```

```
#scaling data
x_train=tf.keras.utils.normalize(x_train,axis=1)
```

```
#scaling data x_test
x_test=tf.keras.utils.normalize(x_test,axis=1)
```

```
#building model #sequential model(feed forward)
model=tf.keras.models.Sequential()
```

```
#add flattern layers i.e not passing 28*28 image passing flatten 
#first layer inpute layer
model.add(tf.keras.layers.Flatten())
```

```
#adding 1st hidden layers
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu)) #we use 128 neurons in layers

#adding 2st hidden layers
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu)) #we use 128 neurons in layers
```

```
#output layer
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))
```

```
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```
model.fit(x_train,y_train,epochs=4)
```

```
#for epoches 4 get minimum loss and more accuracy

val_loss,val_acc=model.evaluate(x_test,y_test)
print(val_loss,val_acc)
```

```
predication=model.predict([x_test])
```
